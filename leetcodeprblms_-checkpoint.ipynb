## Running Sum of 1d Array

```python
class Solution:
    def runningSum(self, nums):
        result = []
        s = 0
        for n in nums:
            s += n
            result.append(s)
        return result



```

```python
nums = [1, 2, 3, 4]
```

```python
obj = Solution()
print(obj.runningSum([1, 2, 3, 4]))
```

```python
nums = [3, 1, 2, 10, 1]
```

```python
obj = Solution()
print(obj.runningSum([3, 1, 2, 10, 1]))
```

```python
class Solution:
    def shuffle(self, nums: list[int], n: int) -> list[int]:
        result = []
        
        for i in range(n):
            result.append(nums[i])
            result.append(nums[i + n])
        
        return result
```

```python
nums = [2,5,1,3,4,7]
n = 3
```

```python
obj1 = Solution()
print(obj1.shuffle(nums,n))
```

```python
nums = [1,2,3,4,4,3,2,1]
n = 4
```

```python
obj1 = Solution()
print(obj1.shuffle(nums,n))
```

```python
class Solution:
    def kidsWithCandies(self, candies: list[int], extraCandies: int) -> list[bool]:
        max_candies = max(candies)
        result = []

        for c in candies:
            if c + extraCandies >= max_candies:
                result.append(True)
            else:
                result.append(False)

        return result
```

```python
obj2 = Solution()
print(obj2.kidsWithCandies(candies = [2,3,5,1,3],extraCandies = 3))
print(obj2.kidsWithCandies(candies = [4,2,1,1,2],extraCandies = 1))
print(obj2.kidsWithCandies(candies = [12,1,12],extraCandies = 10))
```
