```python
class Solution:
    def defangIPaddr(self, address: str) -> str:
        result = ""
        for ch in address:
            if ch == ".":
                result += "[.]"
            else:
                result += ch
        return result
```

```python
obj = Solution()
print(obj.defangIPaddr("255.100.50.0"))
```

```python
from typing import List

class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        count = 0
        for n in nums:
            if len(str(n)) % 2 == 0:
                count += 1
        return count

obj1 = Solution()
print(obj1.findNumbers([12,345,2,6,7896]))
print(obj1.findNumbers([555,901,482,1771]))
```

```python
from typing import List

class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        freq = {}
        count = 0

        for num in nums:
            if num in freq:
                count += freq[num]
                freq[num] += 1
            else:
                freq[num] = 1

        return count
```

```python
obj2 = Solution()
print(obj2.numIdenticalPairs([1,2,3,1,1,3]))
print(obj2.numIdenticalPairs([1,1,1,1]))
print(obj2.numIdenticalPairs([1,2,3]))
```

```python
from typing import List

class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        sorted_nums = sorted(nums)
        count_map = {}

        for i, num in enumerate(sorted_nums):
            if num not in count_map:
                count_map[num] = i

        return [count_map[num] for num in nums]
```

```python
obj3 = Solution()
print(obj3.smallerNumbersThanCurrent([8,1,2,2,3]))
print(obj3.smallerNumbersThanCurrent([6,5,4,8]))
print(obj3.smallerNumbersThanCurrent([7,7,7,7]))
```

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        result = 0
        for i in range(n):
            result ^= (start + 2 * i)
        return result
```

```python

obj = Solution()
print(obj.xorOperation(4, 3)) 
print(obj.xorOperation(5, 0))
```

```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        product = 1
        total = 0

        while n > 0:
            digit = n % 10
            product *= digit
            total += digit
            n //= 10

        return product - total
```

```python
obj = Solution()
print(obj.subtractProductAndSum(234))   # 15
print(obj.subtractProductAndSum(4421))  # 21
```
